```python
# ================================
# Adaptive Risk Index - CKD Full Pipeline (Final Clean Version)
# Run in Google Colab
# ================================

# --------------------------
# 0. Installs & Imports
# --------------------------
!pip install shap --quiet

import warnings
warnings.filterwarnings("ignore")

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import MinMaxScaler, LabelEncoder, LabelBinarizer
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, accuracy_score, f1_score, confusion_matrix, roc_curve
from sklearn.cluster import KMeans

import shap
shap.initjs()

RANDOM_STATE = 42

# --------------------------
# 1. Mount Google Drive (Colab)
# --------------------------
from google.colab import drive
drive.mount('/content/drive')

# --------------------------
# 2. Load Dataset
# --------------------------
file_path = '/content/drive/MyDrive/Colab Notebooks/CKD_project.csv'  # change if needed
df = pd.read_csv(file_path)

print("Initial shape:", df.shape)
print("Columns & Data Types:\n", df.dtypes)

# --------------------------
# 3. Data Cleaning
# --------------------------
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

# Fill NaNs
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].mean())
for col in categorical_cols:
    if df[col].isna().any():
        df[col] = df[col].fillna(df[col].mode()[0])

df.drop_duplicates(inplace=True)
print("After cleaning:", df.shape)

# --------------------------
# 4. Outlier Handling
# --------------------------
def cap_outliers(series):
    Q1, Q3 = series.quantile([0.25, 0.75])
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5*IQR, Q3 + 1.5*IQR
    return np.clip(series, lower, upper)

for col in numeric_cols:
    if df[col].nunique() > 10:
        df[col] = cap_outliers(df[col])
print("Outliers capped ✅")

# --------------------------
# 5. Encoding Categorical Columns
# --------------------------
encoders = {}
for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    encoders[col] = le
print("Categorical columns encoded ✅")

# --------------------------
# 6. Scaling Numeric Columns
# --------------------------
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
scaler = MinMaxScaler()
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])
print("Numeric columns scaled ✅")

# --------------------------
# 7. Feature Engineering
# --------------------------
if "GFR" in df.columns:
    df["GFR_bins"] = pd.cut(
        df["GFR"], bins=[-0.001,0.25,0.5,0.75,1.0],
        labels=["Severe","Moderate","Mild","Normal"]
    ).astype(str)
    df["GFR_bins"] = LabelEncoder().fit_transform(df["GFR_bins"])
    print("GFR discretized ✅")

# PCA
pca_df = None
numeric_for_pca = [c for c in numeric_cols if c not in ['PatientID']]
if len(numeric_for_pca) > 5:
    pca = PCA(n_components=5, random_state=RANDOM_STATE)
    pca_data = pca.fit_transform(df[numeric_for_pca])
    pca_df = pd.DataFrame(pca_data, columns=[f"PCA{i+1}" for i in range(5)])
    print("PCA done ✅")

# --------------------------
# 8. Adaptive Weighting
# --------------------------
key_features = ["GFR","SerumCreatinine","BUNLevels","HemoglobinLevels"]
key_features = [f for f in key_features if f in df.columns]

if len(key_features) == 0:
    raise ValueError("No key features found!")

adaptive_weights = {f: 1.0 for f in key_features}
if "Diagnosis" in df.columns:
    if df["Diagnosis"].dtype == 'object':
        df["Diagnosis"] = LabelEncoder().fit_transform(df["Diagnosis"].astype(str))
    X_rf, y_rf = df[key_features], df["Diagnosis"]
    rf = RandomForestClassifier(n_estimators=100, random_state=RANDOM_STATE)
    rf.fit(X_rf, y_rf)
    importances = rf.feature_importances_
    adaptive_weights = {f: w for f,w in zip(key_features, importances)}

total_w = sum(adaptive_weights.values())
adaptive_weights = {k: v/total_w for k,v in adaptive_weights.items()}
print("Adaptive Weights:", adaptive_weights)

# --------------------------
# 9. Risk Index
# --------------------------
def compute_risk_index(row, weights):
    return sum(row[f]*weights.get(f,0) for f in weights)

df["Risk_Index"] = df.apply(lambda x: compute_risk_index(x, adaptive_weights), axis=1)
df["Risk_Category"] = df["Risk_Index"].apply(
    lambda s: "Low" if s<=0.33 else ("Moderate" if s<=0.66 else "High")
)
print(df[["Risk_Index","Risk_Category"]].head())

# --------------------------
# 10. Visualization
# --------------------------
plt.figure(figsize=(8,5))
sns.histplot(df["Risk_Index"], bins=20, kde=True)
plt.title("Distribution of CKD Risk Index")
plt.show()

weights_df = pd.DataFrame({"Feature":list(adaptive_weights.keys()),"Weight":list(adaptive_weights.values())})
sns.barplot(data=weights_df, x="Feature", y="Weight", hue="Feature", dodge=False)
plt.legend([],[],frameon=False)
plt.title("Adaptive Feature Importance")
plt.show()

# --------------------------
# 11. Patient Clustering
# --------------------------
X_cluster = df[key_features].fillna(0)
kmeans = KMeans(n_clusters=3, random_state=RANDOM_STATE)
df["Cluster"] = kmeans.fit_predict(X_cluster)

sns.scatterplot(x=df[key_features[0]], y=df[key_features[1]], hue=df["Cluster"], palette="Set2")
plt.title("Patient Clusters")
plt.show()

# --------------------------
# 12. SHAP Explainability
# --------------------------
try:
    explainer = shap.TreeExplainer(rf)
    shap_values = explainer.shap_values(df[key_features])

    if isinstance(shap_values, list):
        print(f"SHAP computed for {len(shap_values)} classes.")
        shap.summary_plot(shap_values[0], df[key_features], plot_type="bar", show=True)
    else:
        shap.summary_plot(shap_values, df[key_features], plot_type="bar", show=True)

    patient_idx = 0
    display(shap.force_plot(explainer.expected_value[0], shap_values[0][patient_idx], df[key_features].iloc[patient_idx]))
except Exception as e:
    print("SHAP skipped:", e)

# --------------------------
# 13. Evaluation
# --------------------------
if "Diagnosis" in df.columns:
    X, y = df[key_features], df["Diagnosis"]
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=RANDOM_STATE,stratify=y)
    risk_test = X_test.apply(lambda x: compute_risk_index(x, adaptive_weights), axis=1)
    lb = LabelBinarizer()
    y_test_bin = lb.fit_transform(y_test)

    try:
        if y_test_bin.ndim==1 or y_test_bin.shape[1]==1:
            auc = roc_auc_score(y_test_bin.ravel(), risk_test)
        else:
            aucs=[]
            for c in range(y_test_bin.shape[1]):
                aucs.append(roc_auc_score(y_test_bin[:,c], risk_test))
            auc = np.mean(aucs)
        print("ROC-AUC:", round(auc,3))
    except: pass

# --------------------------
# 14. Save Outputs
# --------------------------
df.to_csv("/content/drive/MyDrive/Adaptive_CKD_Refined.csv", index=False)
print("Refined dataset saved ✅")

# --------------------------
# 15. Summary of Results
# --------------------------
print("\n===== CKD Risk Assessment Summary =====\n")

# Risk category counts
risk_counts = df["Risk_Category"].value_counts()
print("Number of patients in each Risk Category:")
print(risk_counts, "\n")

# Mean and median risk index
print("Risk Index statistics:")
print(f"Mean Risk Index: {df['Risk_Index'].mean():.3f}")
print(f"Median Risk Index: {df['Risk_Index'].median():.3f}\n")

# Adaptive weights
print("Adaptive feature weights:")
for f, w in adaptive_weights.items():
    print(f"{f}: {w:.3f}")
print("")

# Cluster summary
if "Cluster" in df.columns:
    cluster_counts = df["Cluster"].value_counts().sort_index()
    print("Patient Cluster distribution:")
    for c, count in cluster_counts.items():
        print(f"Cluster {c}: {count} patients")
    print("")

# ROC-AUC if Diagnosis exists
if "Diagnosis" in df.columns:
    try:
        if y_test_bin.ndim==1 or y_test_bin.shape[1]==1:
            auc = roc_auc_score(y_test_bin.ravel(), risk_test)
        else:
            aucs=[]
            for c in range(y_test_bin.shape[1]):
                aucs.append(roc_auc_score(y_test_bin[:,c], risk_test))
            auc = np.mean(aucs)
        print(f"ROC-AUC of Risk Index vs Diagnosis: {auc:.3f}")
    except:
        print("ROC-AUC calculation failed")
print("\n===== End of Summary =====\n")
```
